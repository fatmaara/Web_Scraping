## 2. Terjemahkan proses yang ada pada code `extract_usd` tersebut apa maknanya dan tulis dalam markdown ?

```python
# Konversi angka ke satuan USD
import numpy as np

def extract_usd(value):

    # Cek jika dalam USD

    if "USD" in value or "$" in value:

        usd_match = re.search(r'(\d+[\.,]?\d*)', value)

        if usd_match:

            return float(usd_match.group(1).replace(',', ''))

    # Cek jika dalam Rupiah

    elif "Rp" in value:

        rp_match = re.search(r'Rp([\d\.]+)', value)

        if rp_match:

            rupiah = float(rp_match.group(1).replace('.', ''))

            return np.ceil(rupiah / 16380)  # Konversi ke USD dan bulatkan ke atas

    return np.nan  # Jika tidak sesuai format
```

```python
# Tambahkan kolom baru untuk nilai USD
# df_villa['harga dalam USD'] = df_villa['harga sewa ($ USD)'].apply(extract_usd)
# atau memakai lambda
df_villa['harga dalam USD'] = df_villa['harga sewa ($ USD)'].apply(lambda x: extract_usd(x))
df_villa
```

Proses yang ada dalam kode extract_usd berfungsi untuk mengonversi nilai yang ada pada kolom "harga sewa ($ USD)" menjadi nilai dalam satuan USD. Berikut adalah penjelasan dari proses yang ada di dalam kode tersebut:

Penjelasan Proses extract_usd
Pengecekan Format Nilai: Fungsi pertama-tama memeriksa apakah nilai yang diberikan mengandung informasi dalam bentuk USD atau Rupiah. Jika mengandung kata "USD" atau "$", maka nilai dianggap dalam USD. Jika mengandung "Rp", maka dianggap dalam Rupiah.

Ekstraksi Nilai Angka:

Untuk USD: Jika nilai mengandung "USD" atau "$", kode akan menggunakan ekspresi reguler (re.search(r'(\d+[\.,]?\d*)', value)) untuk mencari angka yang ada di dalam string tersebut, termasuk angka dengan koma atau titik sebagai pemisah ribuan.
Untuk Rupiah: Jika nilai mengandung "Rp", kode mencari pola angka setelah "Rp" dengan ekspresi reguler (re.search(r'Rp([\d\.]+)', value)). Kemudian, angka tersebut akan diubah menjadi nilai float, menghapus titik yang ada (jika ada).
Konversi Rupiah ke USD:

Jika nilai dalam Rupiah ditemukan, nilai tersebut akan dibagi dengan kurs 16.380 (kurs yang digunakan dalam kode) untuk dikonversi menjadi USD. Setelah itu, nilai akan dibulatkan ke atas menggunakan fungsi np.ceil().
Jika Tidak Ada Format yang Dikenali:

Jika nilai tidak mengandung informasi yang jelas tentang format uang, maka fungsi akan mengembalikan np.nan sebagai hasil.
Penerapan Fungsi pada Kolom: Fungsi extract_usd diterapkan pada kolom "harga sewa ($ USD)" di dataframe dengan menggunakan apply() atau lambda function. Hasil konversi ini akan disimpan dalam kolom baru yang bernama "harga dalam USD".

## 4. Jelaskan apa makna proses code pada function `convert_to_float` tulis dalam markdown ?

```python
# Konversi kolom 'kamar' dan 'jumlah_kamar_mandi' ke float

def convert_to_float(value):

    try:

        return float(value)

    except ValueError:

        return np.nan  # Jika tidak dapat dikonversi, isi dengan NaN



df_villa['kamar'] = df_villa['kamar'].apply(convert_to_float)

df_villa['jumlah_kamar_mandi'] = df_villa['jumlah_kamar_mandi'].apply(convert_to_float)
df_villa.info()
```


Fungsi convert_to_float digunakan untuk mengonversi nilai pada kolom kamar dan jumlah_kamar_mandi menjadi tipe data float. Berikut adalah penjelasan lebih rinci mengenai proses kode ini:

Fungsi convert_to_float:

Fungsi ini menerima satu parameter value yang akan diproses.
try block: Mencoba mengonversi nilai value menjadi tipe data float.
except block: Jika terjadi kesalahan dalam konversi (misalnya, jika nilai tersebut bukan angka atau berisi teks), maka akan mengembalikan nilai np.nan (Not a Number) untuk menunjukkan bahwa nilai tersebut tidak dapat dikonversi menjadi float.
Proses pada kolom kamar dan jumlah_kamar_mandi:

df_villa['kamar'].apply(convert_to_float) menerapkan fungsi convert_to_float pada setiap elemen dalam kolom kamar, mengonversinya ke tipe data float.
Begitu pula dengan kolom jumlah_kamar_mandi, yang juga diterapkan fungsi yang sama untuk mengonversi nilainya ke tipe float.
